```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def norm(x, u, var):
    return np.exp(-(x-u)**2/(2*var))/np.sqrt(2*np.pi*var)

def plot(
        t,
        u_1, var_1,
        u_2, var_2,
        s_clip = 0,
        e_clip = -1,
        happy_prior = 0.5
    ):
    sad_prior = 1 - happy_prior
    
    plt.title("Posterior")
    happy_plot, = plt.plot(t, norm(t, u_1, var_1) * happy_prior, label="happy")
    sad_plot, = plt.plot(t, norm(t, u_2, var_2) * sad_prior, label="sad")
    plt.legend([happy_plot, sad_plot], ['happy', 'sad'])
    plt.show()

    happy_plot, = plt.plot(t[s_clip:e_clip], norm(t[s_clip:e_clip], u_1, var_1) * happy_prior)
    sad_plot, = plt.plot(t[s_clip:e_clip], norm(t[s_clip:e_clip], u_2, var_2) * sad_prior)
    plt.legend([happy_plot, sad_plot], ['happy', 'sad'])
    plt.title("Posterior (Zoom)")
    plt.grid(True)
    plt.show()
    
t = np.arange(-5.0, 10.0, 0.1)
plot(
    t, s_clip = 50, e_clip = 100,
    u_1 = 5, var_1 = 2,
    u_2 = 0, var_2 = 2
)
```
